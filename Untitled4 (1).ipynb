```python
#remove the outlair
outliers_indexes = []
target = 'Selling_Price'

for col in df.select_dtypes(include='object').columns:
    for cat in df[col].unique():
        df1 = df[df[col] == cat]
        q1 = df1[target].quantile(0.25)
        q3 = df1[target].quantile(0.75)
        iqr = q3-q1
        maximum = q3 + (1.5 * iqr)
        minimum = q1 - (1.5 * iqr)
        outlier_samples = df1[(df1[target] < minimum) | (df1[target] > maximum)]
        outliers_indexes.extend(outlier_samples.index.tolist())


for col in df.select_dtypes(exclude='object').columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1
    maximum = q3 + (1.5 * iqr)
    minimum = q1 - (1.5 * iqr)
    outlier_samples = df[(df[col] < minimum) | (df[col] > maximum)]
    outliers_indexes.extend(outlier_samples.index.tolist())

outliers_indexes = list(set(outliers_indexes))
print('{} outliers were identified, whose indices are:\n\n{}'.format(len(outliers_indexes), outliers_indexes))
```

```python
# Outliers Labeling
df1 = df.copy()
df1['label'] = 'Normal'
df1.loc[outliers_indexes,'label'] = 'Outlier'

# Removing Outliers
removing_indexes = []
removing_indexes.extend(df1[df1[target]>33].index)
removing_indexes.extend(df1[df1['Kms_Driven']>400000].index)
df1.loc[removing_indexes,'label'] = 'Removing'

# Plot
target = 'Selling_Price'
features = df.columns.drop(target)
colors = ['#0055ff','#ff7000','#23bf00']
CustomPalette = sns.set_palette(sns.color_palette(colors))
fig, ax = plt.subplots(nrows=3 ,ncols=3, figsize=(15,12), dpi=200)

for i in range(len(features)):
    x=i//3
    y=i%3
    sns.scatterplot(data=df1, x=features[i], y=target, hue='label', ax=ax[x,y])
    ax[x,y].set_title('{} vs. {}'.format(target, features[i]), size = 15)
    ax[x,y].set_xlabel(features[i], size = 12)
    ax[x,y].set_ylabel(target, size = 12)
    ax[x,y].grid()

ax[2, 1].axis('off')
ax[2, 2].axis('off')
plt.tight_layout()
plt.show()
```

```python
#spliting the target feture out of the independent fetures
x= df1.drop('Selling_Price', axis=1)
y = df1['Selling_Price']
```

```python
#usage of split function

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=0)

print('xtrain shape: ', xtrain.shape)
print('xtest shape: ', xtest.shape)
print('ytrain shape: ', ytrain.shape)
print('ytest shape: ',ytest.shape)
```

```python
#Scale Data using Standard Scaler

scaler = StandardScaler()
scaler.fit(xtrain)
xtrain_scaled = scaler.transform(xtrain)
xtest_scaled = scaler.transform(xtest)
```

# Train the Model

```python
linear_reg = LinearRegression()
linear_reg.fit(xtrain_scaled, ytrain)
```

```python
pd.DataFrame(data = np.append(linear_reg.intercept_ , linear_reg.coef_),
             index = ['Intercept']+[col+" Coef." for col in x.columns], columns=['Value']).sort_values('Value', ascending=False)
```

```python
#evaluation
def model_evaluation(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    MAE = metrics.mean_absolute_error(y_test, y_pred)
    MSE = metrics.mean_squared_error(y_test, y_pred)
    RMSE = np.sqrt(MSE)
    R2_Score = metrics.r2_score(y_test, y_pred)

    return pd.DataFrame([MAE, MSE, RMSE, R2_Score], index=['MAE', 'MSE', 'RMSE' ,'R2-Score'], columns=[model_name])
```

```python
#evaluation
from sklearn import metrics


def model_evaluation(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    MAE = metrics.mean_absolute_error(y_test, y_pred)
    MSE = metrics.mean_squared_error(y_test, y_pred)
    RMSE = np.sqrt(MSE)
    R2_Score = metrics.r2_score(y_test, y_pred)

    return pd.DataFrame([MAE, MSE, RMSE, R2_Score], index=['MAE', 'MSE', 'RMSE', 'R2-Score'], columns=[model_name])

# Evaluate the Random Forest model
rf_evaluation = model_evaluation(rf_model, xtest_scaled, ytest, 'Random Forest')

# Display the evaluation metrics
print(rf_evaluation)
```

# the evaluation compare

```python
import matplotlib.pyplot as plt

# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(xtrain_scaled, ytrain)
linear_reg_evaluation = model_evaluation(linear_reg, xtest_scaled, ytest, 'Linear Regression')

# Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(xtrain_scaled, ytrain)
rf_evaluation = model_evaluation(rf_model, xtest_scaled, ytest, 'Random Forest')

# Combine evaluation results
combined_evaluation = pd.concat([linear_reg_evaluation, rf_evaluation], axis=1)

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# MAE Plot
axes[0, 0].bar(combined_evaluation.columns, combined_evaluation.loc['MAE'])
axes[0, 0].set_title('Mean Absolute Error (MAE)')

# MSE Plot
axes[0, 1].bar(combined_evaluation.columns, combined_evaluation.loc['MSE'])
axes[0, 1].set_title('Mean Squared Error (MSE)')

# RMSE Plot
axes[1, 0].bar(combined_evaluation.columns, combined_evaluation.loc['RMSE'])
axes[1, 0].set_title('Root Mean Squared Error (RMSE)')

# R2-Score Plot
axes[1, 1].bar(combined_evaluation.columns, combined_evaluation.loc['R2-Score'])
axes[1, 1].set_title('R2-Score')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
